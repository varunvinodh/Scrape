## Creating the Function

```python
# This function was created on 04-09-2021

def get_data(url):
    # Setting path for Gekodriver
    import os
    path=r'C:\Users\varun\Desktop\interv data\Scrape'
    os.chdir(path)
    
    #Importing usefull packages
    import pandas as pd
    from selenium import webdriver
    from bs4 import BeautifulSoup
    
    # Enter the URL of the Videos section of a YouTube Channel
    # url=input("Enter the url of the video's page :")
    
    browser=webdriver.Firefox(path)
    browser.get(url)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    
    titles = soup.findAll('a',id='video-title')
    video_urls = soup.findAll('a',id='video-title')
    views_release =soup.findAll('div',id='metadata-line')

    
    title=[]
    released_on,views=[],[]
    video_url=[]

    for i in range(len(views_release)):
        title.append(titles[i].text)
    
        sep = 'views'
        stripped=[]
    
        stripped.append(views_release[i].text.split(sep, 1))
        views.append(stripped[0][0])
        released_on.append(stripped[0][1])
    
        video_url.append('https://www.youtube.com'+video_urls[i].get('href'))
    
    # Creating a DataFrame out of the scraped data
    dict_ = {'Title': title, 'Views': views, 'Date': released_on, 'Link': video_url} 
    df = pd.DataFrame(dict_)
    
    return(df)
```

## Calling the Function

```python
# Amazon India Videos link.
url='https://www.youtube.com/c/AmazonInOfficial/videos'
url
```

```python
get_data(url)
```

## Saving the Dataframe to work on it

```python
df=get_data(url)
```
